```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get('http://info.cern.ch/hypertext/WWW/TheProject.html')
html=response.text
# print(response.content.decode('UTF-8'))
```

```python
data=BeautifulSoup(html,'html.parser')
print(data.prettify())
```

```python

print(data.get_text())
```

```python
li=data.find_all()
for i in li:
    print(list(i.strings))
```
